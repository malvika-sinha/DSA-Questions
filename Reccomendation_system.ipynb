##POPULARITY BASED RECOMMENDATION SYSTEM

```python
#.merge() is a method provided by data manipulation libraries like pandas in Python. It is used to combine two data tables based on a common column.
#In this case, the common column is specified as 'ISBN'
myratings = ratings.merge(books,on='ISBN')
```

```python
ratings.merge(books,on='ISBN').shape#no if records after merging(it has 44461 rows and 10 columns)
```

```python
myratings.groupby('Book-Title').count()['Book-Rating']
```

```python
numrating = myratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
numrating.rename(columns={'Book-Rating':'numrating'},inplace= True)
numrating
```
